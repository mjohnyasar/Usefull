```python

import docx

doc = docx.Document("sql.docx")
```

```python
all_paras = doc.paragraphs
len(all_paras)
```

```python
for para in all_paras:
    print(para.text)
```

```python

Tr2Eng = str.maketrans("İıÜüÖöÇçŞşĞğ","IiUuOoCcSsGg")
```

```python
for i in range(len(all_paras)):
    ma = doc.paragraphs[i]
    runTr=ma.text
    runEng = runTr.translate(Tr2Eng)
    print(runEng)
```

```python
mydoc = docx.Document()
```

```python
for i in range(len(all_paras)):
    ma = doc.paragraphs[i]
    runTr=ma.text
    runEng = runTr.translate(Tr2Eng)
    mydoc.add_paragraph(runEng)
mydoc.save("mydoc.docx")
print("mydoc.docx  saved")   
```
